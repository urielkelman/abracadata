```python
def esta_en_rango(x, limit_inf, limit_sup):
    if x >= limit_inf and x < limit_sup:
        return True
    return False
```

```python
postulantes['35_a_50'] = postulantes['edad'].apply(lambda x: esta_en_rango(x, 3, 51 ))
```

```python
postulantes['mayor_de_50'] = postulantes['edad'].apply(lambda x: esta_en_rango(x, 51, 250 ))
```

```python
def ventas(x):
    return True if x == "Ventas" else False
def administracion(x):
    return True if x == "Administración" else False
def produccion(x):
    return True if x == "Producción" else False
def comercial(x):
    return True if x == "Comercial" else False
def at_cliente(x):
    return True if x == "Atención al Cliente" else False
def recep(x):
    return True if x == "Recepcionista" else False
def call_center(x):
    return True if x == "Call Center" else False
def cont(x):
    return True if x == "Contabilidad" else False
def program(x):
    return True if x == "Programación" else False
def tecno(x):
    return True if x == "Tecnologia / Sistemas" else False
def telemark(x):
    return True if x == "Telemarketing" else False
def teso(x):
    return True if x == "Tesorería" else False
def almacen(x):
    return True if x == "Almacén / Depósito / Expedición" else False
def rrhh(x):
    return True if x == "Recursos Humanos" else False
def limpieza(x):
    return 1 if x == "Mantenimiento y Limpieza" else 0
```
