```python
postulantes = pd.read_csv("Data/fiuba_entrenamiento/pocho/postulantes.csv")
postulantes.head()
```

```python
postulantes.edad.mean()
```

```python
intento= pd.get_dummies(postulantes[['sexo','Doctorado','Master','Posgrado','Secundario','Terciario/Técnico','Universitario']])
```

```python
intento
```

```python
intento['sexo_0.0'].sum()
```

```python
intento.drop(['sexo_0.0','Doctorado_-','Master_-','Posgrado_-','Secundario_-','Terciario/Técnico_-','Universitario_-'],axis=1,inplace=True)
intento.head()
```

```python
intento.dtypes
```

```python
for column in intento:
    intento[column]=(intento[column]==1)
intento.dtypes
```

```python
intento.head()
```

```python
merge=pd.merge(postulantes,intento, left_index=True, right_index=True)
merge.head()
```

```python
merge.drop(['sexo','Doctorado','Master','Otro','Posgrado','Secundario','Terciario/Técnico','Universitario'],axis=1,inplace=True)
merge.head()
```

```python
merge.to_csv("Data/fiuba_entrenamiento/pocho/postulantesONEHOT.csv",index=False)
```
