```python
postulaciones = pd.read_csv("../../Data/fiuba_entrenamiento/gian/postulaciones.csv")
postulaciones['sepostulo']=1
postulaciones.drop(['fechapostulacion','dia','mes','semana','diadelasemana','hora'],axis=1,inplace=True)
postulaciones.head()
```

```python
postulantes = pd.read_csv("postulantes_amp.csv")
postulantes.head()
```

```python
postulaciones['sepostulo'] = 1
```

```python
no_postulaciones = pd.read_csv("../../Data/fiuba_entrenamiento/gian/no-postulaciones/8.csv")
```

```python
postulaciones=postulaciones.append(no_postulaciones, ignore_index=True)
postulaciones.drop_duplicates(['idaviso','idpostulante'],keep='first',inplace=True)
print(len(postulaciones))
postulaciones.head()
```

```python
postulaciones = postulaciones.sample(n=6000000)
```

```python
len(postulaciones)
```

```python
postulaciones=pd.merge(postulaciones,postulantes,on='idpostulante')
```

```python
postulaciones=pd.merge(postulaciones,avisos,on='idaviso')
```

```python
postulaciones.head()
```

```python
vistas.head()
```

```python
postulaciones=pd.merge(postulaciones,vistas,on=['idaviso','idpostulante'],how='left')
```

```python
postulaciones['vio_anuncio'] = postulaciones['vio_anuncio'].fillna(0)
postulaciones['cantidad'] = postulaciones['cantidad'].fillna(0)

postulaciones.sample(n=15)
```

```python
train, test=train_test_split(postulaciones,test_size=0.10)
print("Train: ",len(train),"Test: ",len(test))
```

```python
features=list(postulaciones.columns)
features
```

```python
features.remove('idaviso')
features.remove('idpostulante')
features.remove('sepostulo')
features.remove('Unnamed: 0')
features
```

```python
features.remove('titulo')
features.remove('denominacion_empresa')
```

```python
x_train=train[features]
y_train=train['sepostulo']

x_test=test[features]
y_test=test['sepostulo']
```

```python
c=lgb.LGBMClassifier(learning_rate=0.5,objective='binary', max_depth=20, num_leaves=3000,n_estimators=50,colsample_bytree=0.9,n_jobs=-1,random_state=0,silent=False,subsample=1,sumsample_freq=1, two_round=True)

dt=c.fit(x_train,y_train,eval_set=[(x_test,y_test)],verbose=True,early_stopping_rounds=1)
```

```python
score=c.score(x_test,y_test)*100
print(score)
```

```python
joblib.dump(c, "../../Data/fiuba_entrenamiento/gian/modelos/LIGTHBM(learning-rate:0.5).pkl")
```
